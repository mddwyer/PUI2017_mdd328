## Question: 
Is the new Bus route improving commute?

## Null Hypothesis:
The new average Bus route time is the same or higher than the old average Bus route time.

$H_0$ : $\ New Bus Time Mean >= Old Bus Time Mean$

## Alternative Hypothesis:
The new average Bus route time is significantly lower than the old average Bus route time. 

$H_1$ : $\ New Bus Time Mean < Old Bus Time Mean$

The significance level will be $\alpha=0.05$.

```python
# New Bus Times
tNew = pd.read_csv("https://raw.githubusercontent.com/fedhere/PUI2017_fb55/"
             + "master/Lab3_fb55/times.txt",header=None).values.squeeze()
tNew
```

```python
tNewMean = tNew.mean()
```

```python
tOldMean = 36.
tOldStdev = 6.
```

```python
z_score = (tOldMean - tNewMean) / (tOldStdev / (sqrt(100)))
z_score
```

The null hypothesis can be rejected because the calculated Z score reveals the new bus times being over 2 standard deviations from the mean, greater than the stated 0.05 (2 standard deviations) significance level.
